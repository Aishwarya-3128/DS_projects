```python
# import modules
import pandas as pd
from  matplotlib import pyplot as plt
import seaborn as sns
```

```python
# read dataset
df_data= pd.read_csv('Online_shopping.csv');
df_data.head()
```

```python
# dataset size
x,y=df_data.shape
print('There are {0} rows and {1} columns.'.format(x,y))
```

```python
# To get the columns in dataset
df_data.columns
```

```python
# To get the information about the dataset
df_data.info()
```

```python
# To get statistical information about the numerical columns
df_data.describe()
```

```python
# to check if the dataset have missing values
df_data.isnull().sum()
```

```python
## dropping rows with null customer id
df_data.dropna(subset=['CustomerID'],inplace= True)
```

```python
### checking if there are customer id with missing values
## Ignoring coupon code and discount since there wont be coupon or discount applied all times
print(df_data.isnull().sum())
```

```python
## To find purchase pattern of the customer
cust_purchase_patterns = df_data.groupby('CustomerID')['Product_Category'].nunique()
print(cust_purchase_patterns)
```

```python
sales_by_category = df_data.groupby('Product_Category')['Quantity'].sum()
print(sales_by_category.idxmax())
```

## Univariant Visualization

```python
## Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data =df_data, x = 'Avg_Price', kde = True)
plt.title('Histogram of Average price')  
plt.show()
```

```python
## Pie chart
plt.pie(df_data['Gender'].value_counts(),labels=['Female','Male'],autopct='%1.1f%%')
plt.title('Gender Distribution')  
plt.show()
```

```python
## Count plot
plt.figure(figsize=(10, 10))
sns.countplot(x ='Product_Category', data = df_data)
plt.title('Product category counts')  
plt.xlabel('Product category')
plt.ylabel('Count of Product')
plt.xticks(rotation=90)
plt.show()
```

## Bivariant visualization

```python
## Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Product_Category', y='Online_Spend', data=df_data)
plt.title('Online Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Online spend')
plt.xticks(rotation=90)
plt.show()
```

```python
## Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Avg_Price', data=df_data)
plt.title('Product Quantity vs Average Price')
plt.xlabel('Quantity')
plt.ylabel('Average Price')
plt.show()
```

```python
## Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Month',y='Offline_Spend',data=df_data)
plt.xlabel('Month number')
plt.ylabel('Offline sales')
plt.show()
```

## Multivariant Visualization

```python
## Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_data.corr(), annot = True)
plt.show()
```

```python
## Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avg_Price',y='Delivery_Charges',hue='Location',data=df_data)
plt.show()
```

```python
## Pair plot
plt.figure(figsize=(10,6))
sns.pairplot(df_data[['Avg_Price','Offline_Spend','Online_Spend']])
plt.show()
```
