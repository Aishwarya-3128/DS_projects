```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
df.drop('Genre',axis=1,inplace=True)
```

```python
sns.pairplot(df)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sc =  StandardScaler()
```

```python
df_sc = sc.fit_transform(df)
```

```python
display(df_sc)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4,random_state=30).fit(df_sc)
```

```python
km.inertia_
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
df['Labels']=km.labels_
```

```python
df.head()
```

```python
sns.scatterplot(data=df,y='Annual_Income_(k$)',x='Age',hue='Labels')
```

## 5 cluster

```python
km = KMeans(n_clusters=5,random_state=30).fit(df_sc)
```

```python
df['Labels_5']=km.labels_
```

```python
sns.scatterplot(data=df,y='Annual_Income_(k$)',x='Age',hue='Labels_5')
```

## Cluster 6

```python
km = KMeans(n_clusters=6,random_state=30).fit(df_sc)
```

```python
df['Labels_6']=km.labels_
```

```python
sns.scatterplot(data=df,y='Annual_Income_(k$)',x='Age',hue='Labels_6',palette=['Red','Green','Blue','Black','Pink','Orange'])
```
